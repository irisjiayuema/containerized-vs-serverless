# Logistic Regression

## Not Cached

```python
lr_va_nc_transfer = [264.71, 108.77, 292.28, 108.74, 116.8, 260.54]
lr_va_c_transfer1 = [134.57,143.95,173.51,135.83,136.87,144.18]
lr_va_c_transfer2 = [254.33,140.35,141.55,154.08,147.81,133.77]

lr_cf_nc_transfer = [31.79, 43.07, 30.13, 43.7, 38.13,30.52]
lr_cf_c_transfer1 = [27.22,24.31,119.54,28.3,29.86,28.01]
lr_cf_c_transfer2 = [22.38,25.5,26.29,29.44,25.08,24.65]
```

```python
plt.plot(batch_sizes, lr_va_nc_transfer, label='Vertex AI - Trial 1 (Not Cached)', color='lawngreen')
plt.plot(batch_sizes, lr_va_c_transfer1, label='Vertex AI - Trial 2 (Cached)', color='darkgreen')
plt.plot(batch_sizes, lr_va_c_transfer2, label='Vertex AI - Trial 3 (Cached)', color='darkgreen')
plt.plot(batch_sizes, lr_cf_nc_transfer, label='Cloud Functions - Trial 1 (Not Cached)', color='skyblue')
plt.plot(batch_sizes, lr_cf_c_transfer1, label='Cloud Functions - Trial 2 (Cached)', color='darkblue')
plt.plot(batch_sizes, lr_cf_c_transfer2, label='Cloud Functions - Trial 3 (Cached)', color='darkblue')

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.title("Logistic Regression: Vertex AI vs Cloud Functions Transfer Time")
plt.xlabel("Batch Size")
plt.ylabel("Latency (ms)")
```
